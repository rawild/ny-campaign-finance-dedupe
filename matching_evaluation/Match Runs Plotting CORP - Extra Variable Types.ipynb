```python
%matplotlib inline 
import matplotlib.pyplot as plt
from matplotlib import cm
from numpy import linspace
```

```python
files = [['dedupe_extension/settings_0','settings_0 orig',2],
         ['dedupe_extension/settings_1','settings_1 orig',0],
         ['dedupe_extension/settings_2','settings_2 orig',2],
         ['dedupe_extension/settings_3','settings_3 orig',0],
         ['dedupe_extension/settings_4', 'settings_4 orig',0],
         ['dedupe_extension/settings_CORP_ext_4','settings_4 extra',1],
         ['dedupe_extension/settings_CORP_ext_3','settings_3 extra',1],
         ['dedupe_extension/settings_CORP_ext_2','settings_2 extra',1],
         ['dedupe_extension/settings_CORP_ext_1','settings_1 extra',1],
         ['dedupe_extension/settings_CORP_ext_0','settings_0 extra',1],
        ]
```

```python
start = 0.3
stop = 1.0
number_of_lines= 7
cm_subsection = linspace(start, stop, number_of_lines)
cm_subsection_1 = linspace(.5,stop, 4)
purples = [ cm.Purples(x) for x in cm_subsection ]
reds = [ cm.Reds(x) for x in cm_subsection ]
blues = [ cm.Blues(x) for x in cm_subsection_1 ]
colors = [purples,reds,blues]
```

```python
ratio_sets = []
for i, file in enumerate(files):
    data_cut = data_df[data_df['settings_file'] == file[0]].sort_values('total_donors')
    j = i
    if i > 4:
        j = i-5
    color = colors[file[2]][j]
    ratio_sets.append([data_cut['total_donors'].values, data_cut['donor_cluster_ratio'].values, file[1], color])
```

```python
ratio_sets
```

```python
plt.figure(figsize=(8,8))
for s in ratio_sets:
    plt.plot(s[0],s[1], ms=6, marker='.', label=s[2], color=s[3])
plt.xlabel('Total Donors')
plt.ylabel('Donor to Cluster Ratio')
plt.title('Donor to Cluster Ratio for Organization Donors')
plt.legend()
```
