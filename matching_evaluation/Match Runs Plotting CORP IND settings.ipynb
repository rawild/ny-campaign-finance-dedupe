# Visualize Results of Matching Runs from CORP Runs with IND settings

## Import in data from runs

```python
import pandas as pd
```

```python
data_df = pd.read_csv('corp_match_runs_12_4.csv')
```

```python
data_df['settings_file'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
data_df['settings_num'] = encoder.fit_transform(data_df['settings_file'])
```

```python
data_df
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt
from matplotlib import cm
from numpy import linspace
```

```python
files = ['dedupe_extension/settings_0','dedupe_extension/settings_1',
        'dedupe_extension/settings_2','dedupe_extension/settings_3','dedupe_extension/settings_4',
         'dedupe_extension/settings_IND_0','dedupe_extension/settings_IND_1',
        'dedupe_extension/settings_IND_2','dedupe_extension/settings_IND_3','dedupe_extension/settings_IND_4']
```

```python
start = 0.0
stop = 1.0
number_of_lines= len(files)
cm_subsection = linspace(start, stop, number_of_lines) 
colors = [ cm.bwr(x) for x in cm_subsection ]
```

```python
ratio_sets = []
for i, file in enumerate(files):
    data_cut = data_df[data_df['settings_file'] == file].sort_values('total_donors')
    color = colors[i]
    ratio_sets.append([data_cut['total_donors'].values, data_cut['donor_cluster_ratio'].values, file, color])
```

```python
ratio_sets
```

```python
plt.figure(figsize=(8,8))
for s in ratio_sets:
    plt.plot(s[0],s[1], ms=6, marker='.', label=s[2], color=s[3] )
plt.legend()
```

```python
plt.figure(figsize=(8,8))
for s in ratio_sets:
    plt.plot(s[0],s[1], ms=6, marker='.', label=s[2], color=s[3])
plt.xlabel('total donors')
plt.ylabel('donor to cluster ratio')
plt.legend()
```

```python
avg_size_sets = []
for i, file in enumerate(files):
    data_cut = data_df[data_df['settings_file'] == file].sort_values('total_donors')
    color = colors[i]
    avg_size_sets.append([data_cut['total_donors'].values, data_cut['avg_cluster_size'].values, file, color])
```

```python
plt.figure(figsize=(8,8))
for s in avg_size_sets:
    plt.plot(s[0],s[1], ms=6, marker='.', label=s[2], color=s[3])
plt.xlabel('total donors')
plt.ylabel('avg cluster size')
plt.legend()
```
