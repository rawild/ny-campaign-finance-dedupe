```python
import pandas as pd
```

```python
columns = pd.read_table('EFSRECB.txt',skiprows=8, nrows=31, sep='\s+', engine='python')
column_names = columns["FIELD"]
column_list = list(column_names.values)
this_filing = pd.DataFrame(columns=column_list)
file_results_columns = ["file","errored"]
file_results = pd.DataFrame(columns=file_results_columns)
```

```python
df = pd.read_csv("ALL_REPORTS_fixed.txt",header=None, names=column_list, encoding="latin1")
df_inbound = df[df['TRANSACTION_CODE'].isin(['A','B','C','D','E'])]
df_inbound.shape
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split_filings = StratifiedShuffleSplit(n_splits=1, test_size=0.75, random_state=17)
for train_index, test_index in split_filings.split(df_inbound, df_inbound['TRANSACTION_CODE']):
    filers_train = df_inbound.iloc[train_index,:]
```
